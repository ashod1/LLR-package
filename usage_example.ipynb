```python
import numpy as np
from sklearn.neighbors import KDTree
import pylab as plt
from lwlr import LWLR
```

```python
x_train = np.linspace(-10*3.14, 10*3.14, 1000)
noise = np.random.normal(0, 0.1, x_train.shape)
y_train = np.sin(x_train) + noise

x_query = np.linspace(-10*3.14, 10*3.14, 1000)

model = LWLR(weight_type='inverse_distance')

#tic = time.time()
y_pred = model.predict(x_query[:,None], x_train[:,None], y_train, nn=10, extra_weights=1/noise**2)
#print(f"Time taken: {time.time()-tic} seconds")
```

```python
fig, ax = plt.subplots(figsize=(20,6))

ax.plot(x_train, y_train, 'o', label='Training data', alpha=0.5)
ax.plot(x_query, np.sin(x_query), label='True function', c='k')
ax.plot(x_query, y_pred, '-', label='Predictions')
ax.legend()
#ax.set_xlim(-3.2, -2)
```
